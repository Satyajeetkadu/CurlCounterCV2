```python
import cv2
import mediapipe as mp

# Initialize MediaPipe's Pose model
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

# Load an image (replace with your image file)
image = cv2.imread("/Users/satyajeetkadu/Documents/Sem-9/IP LAB/ipproj/Fit.png")

# Convert the image to RGB (MediaPipe uses RGB format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detect the pose landmarks
results = pose.process(image_rgb)

# Extract shoulder and waist keypoints
shoulder1 = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER]
shoulder2 = results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER]
waist1 = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP]
waist2 = results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP]

# Calculate the shoulder-to-waist ratio
shoulder_to_waist_ratio = (
    ((shoulder1.x - shoulder2.x) ** 2 + (shoulder1.y - shoulder2.y) ** 2) ** 0.5 /
    ((waist1.x - waist2.x) ** 2 + (waist1.y - waist2.y) ** 2) ** 0.5
)

# Display the result on the image
cv2.putText(image, f"Shoulder-to-Waist Ratio: {shoulder_to_waist_ratio:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

# Display the image
cv2.imshow("Shoulder-to-Waist Ratio", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe
mp_pose = mp.solutions.pose

# Load the image
image = cv2.imread('/Users/satyajeetkadu/Documents/Sem-9/IP LAB/ipproj/Screenshot 2023-10-20 at 3.10.57 PM.png')

# Initialize the pose model
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    # Convert the BGR image to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Process the image
    results = pose.process(image_rgb)

    # Check if the detection was successful
    if results.pose_landmarks:
        # Extract landmarks
        landmarks = results.pose_landmarks.landmark

        # Find the outermost shoulder points
        left_shoulder = min(landmarks, key=lambda x: x.x)
        right_shoulder = max(landmarks, key=lambda x: x.x)

        # Waist landmarks
        landmarks_on_waist = [landmarks[mp_pose.PoseLandmark.LEFT_HIP], landmarks[mp_pose.PoseLandmark.RIGHT_HIP]]

        # Find the outermost points on the waist
        left_waist = min(landmarks_on_waist, key=lambda x: x.x)
        right_waist = max(landmarks_on_waist, key=lambda x: x.x)

        # Convert landmark positions to pixel coordinates
        height, width, _ = image.shape
        left_shoulder_x = int(left_shoulder.x * width)
        right_shoulder_x = int(right_shoulder.x * width)

        left_waist_x = int(left_waist.x * width)
        right_waist_x = int(right_waist.x * width)

        # Calculate shoulder-to-waist ratio
        shoulder_width = right_shoulder_x - left_shoulder_x
        waist_width = right_waist_x - left_waist_x
        ratio = shoulder_width / waist_width

        # Classify the image
        if ratio < 0.75:  # Adjust the threshold as needed
            classification = "Unfit"
        else:
            classification = "Fit"

        # Draw the classification on the image
        cv2.putText(image, f"Classification: {classification}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Save the output image with markers and classification
    cv2.imwrite('output_image.jpg', image)

# Display the image with shoulder and waist markers and classification
cv2.imshow('Shoulder and Waist Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
